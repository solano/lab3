# Constante de Amortecimento

```python
#Ajuste
def f(t, a, b, c):
    return a*np.exp(-b*t)+c

time = dat[2]*dat[3]
volt = dat[0]*dat[1]
x = np.linspace(0,0.0012,50)

p0 = sy.array([2,80,0])
param, err = opt.curve_fit(f, time, volt, p0=p0)
err = np.sqrt(np.diag(err))

plt.figure(dpi=100)
plt.plot(time, volt, 'bo', label="Dados", markersize=2)
plt.plot(x, f(x,param[0],param[1],param[2]), 'r', label="ajuste", linewidth=1)
plt.legend()
plt.xlabel("Tempo (s)")  ;  plt.ylabel("Tensão (V)")
plt.title("Curva da Tensão pelo tempo no regime transiente sub-amortecido")
plt.ticklabel_format(style="sci", axis="both", scilimits=(0,0))
plt.show()

print("Parâmetros ajustados: \na = {1} +- {2}, y = {2} +- {3}, b = {4} +- {5}".format(param[0],err[0],param[1],err[1],param[2],err[2]) )
```
