```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
import uncertainties as unc
from uncertainties import ufloat as u
import uncertainties.unumpy as unp

R = 98.7 #ohm, escala de 200 ohm
C = 108.8e-9 #F, escala de 200 nF
L = 45.3e-3 #H, escala de 200 mH
erro_R = 0.5+0.8e-2*R
erro_C = 2e-9+2.5e-2*C
erro_L = 3e-3+2.5e-2*L
R = u(R,erro_R)
C = u(C,erro_C)
L = u(L,erro_L)

fteo = 1/(2*np.pi*unp.sqrt(L*C))
```

```python
# Componente(ddp, dt)
class Componente:
    frequencia = u(2071.4,0.5)
    angular = 2*np.pi*frequencia
    periodo = 1/frequencia
    _frequencia = frequencia.nominal_value
    _angular = angular.nominal_value
    _periodo = periodo.nominal_value

    def __init__(self, amplitude, dt):
        self.amplitude = amplitude
        self.fase = self.angular*dt
        
    def f(self, t):
        return self.amplitude*unp.cos(self._angular*t - self.fase)
    def _f(self,t):
        return unp.nominal_values(self.f(t))
    def errf(self,t):
        return unp.std_devs(self.f(t))
    
    def relativo(self, referencia=0):
        return f"amplitude ({self.amplitude}) V, fase ({self.fase - referencia}) rad"
    def __repr__(self):
        return self.relativo(0)
```

```python
pidt = Componente._periodo/2   # meio período

# dados reais tomados do osciloscópio
Fonte = Componente(u(3.7,0.2), u(0,0))
R_1 =   Componente(u(1.9,0.1), u(-70e-6,10e-6))
L_1 =   Componente(u(6.4,0.2), u(5e-6,1e-6))
C_1 =   Componente(u(2.1,0.1), u(-136e-6,2e-6))

# PROBLEMA: com os dados reais, o indutor tá com comportamento capacitivo
# e o capacitor com comportamento indutivo. Isso não faz sentido.
# Outro problema é que a soma de tensões é diferente de zero.

# Primeira ideia: somar/subtrair pi pra ver se tomamos o suplementar
# (que pode realmente acontecer no experimento). Graficamente equivale
# a uma inversão de sinal. Não faz diferença somar ou subtrair.
#L_1 =   Componente(u(6.4,0.2), u(5e-6,1e-6)    -pidt)
#C_1 =   Componente(u(2.1,0.1), u(-136e-6,2e-6) +pidt)
# Os sinais ficam certos mas as fases ultrapassam pi/2 da corrente, o que não
# deveria ser possível se ignorarmos a reatância do resistor (a frequência é
# baixa). Quantitativamente é o resultado que chega mais perto de 0, mas é
# não-físico.

# Segunda ideia: trocar as fases do indutor e do capacitor (podem ter sido
# registradas errado).
#L_1 =   Componente(u(6.4,0.2), u(136e-6,2e-6))
#C_1 =   Componente(u(2.1,0.1), u(5e-6,1e-6))
# Capacitor agora faz sentido e indutor não. Indutor fica com fase entre pi/2 e pi.
# Fazendo também o óbvio:
#L_1 =   Componente(u(6.4,0.2), u(136e-6,2e-6) -pidt)
# os sinais ficam certos, mas o resultado dá bem longe de 0.

# Terceira ideia: trocar por complementares e colocar os sinais corretos com a mão.
# Isso não tem justificativa, não tem um erro de experimento que possa levar a isso.
# É desespero.
#L_1 =   Componente(u(6.4,0.2), u(5e-6,1e-6)    -pidt/2)
#C_1 =   Componente(u(2.1,0.1), u(-136e-6,2e-6) +pidt/2)
# Também dá bem longe de 0.

# Quarta ideia: trocar sinais das fases relativas medidas (podemos registrá-los errado).
#L_1 =   Componente(u(6.4,0.2), -u(5e-6,1e-6))
#C_1 =   Componente(u(2.1,0.1), -u(-136e-6,2e-6))
# Os dois ficam capacitivos, e mesmo o resultado do capacitor é não-físico porque passa
# de pi/2. O óbvio a tentar é tomar suplementares:
#L_1 =   Componente(u(6.4,0.2), -u(5e-6,1e-6)    -pidt)
#C_1 =   Componente(u(2.1,0.1), -u(-136e-6,2e-6) -pidt)
# Ambos ficam indutivos. Indutor fica não-físico mesmo assim porque
# fica entre -pi e -pi/2.

# Não parece haver erro entre fonte e resistor; a fase relativa entre eles faz sentido.
# É positiva (comportamento capacitivo, como esperado pela reatância) e menor que pi/2.

# INCONCLUSIVO.

soma = lambda x:Fonte.f(x)+R_1.f(x)+L_1.f(x)+C_1.f(x)
_soma = lambda x: unp.nominal_values(soma(x))
erro_soma = lambda x: unp.std_devs(soma(x))
```

```python
print('EM RELAÇÃO À TENSÃO NO RESISTOR (CORRENTE):')
print('Resistor:\t',  R_1.relativo(R_1.fase))
print('Indutor:\t',   L_1.relativo(R_1.fase))
print('Capacitor:\t', C_1.relativo(R_1.fase))
print('Fonte:\t\t',   Fonte.relativo(R_1.fase),'> 0 (comportamento capacitivo)')

print('\nreatância:',Componente.angular*L-1/(Componente.angular*C),'< 0 (comportamento capacitivo era esperado)')
```

```python
N=1
xx = np.linspace(0,Fonte._periodo*N, 150)
start = unp.nominal_values(R_1.fase/R_1.angular)
x = np.linspace(start, start+Componente._periodo*N, 150)
# usar x nas funções, xx no eixo do plot
# (isso translada tudo pra corrente ficar como referência)

plt.figure(dpi=100)
plt.ylabel("Tensão (V)")
plt.xlabel("Tempo (s)")
plt.title("Tensão nos terminais do circuito RLC")
plt.ticklabel_format(style="sci", axis="x", scilimits=(0,0))
plt.plot([xx[0], xx[-1]], [0,0], 'k')
plt.plot(xx, Fonte._f(x), label="Fonte")
plt.plot(xx, R_1._f(x),label="R")
plt.plot(xx, L_1._f(x),label="L")
plt.plot(xx, C_1._f(x),label="C")
plt.plot(xx, _soma(x), '--',color='darkcyan', label="Soma")
plt.fill_between(xx, _soma(x)-2*erro_soma(x), _soma(x)+2*erro_soma(x),
                 color="paleturquoise", label=r"$\pm2\sigma$")
plt.xlim([xx[0], xx[-1]])
ax = plt.subplot(111)
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 1, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```
